```python
import pandas as pd
```

**Bire Bir Birleştirme İşlemi**

```python
sozluk1=pd.DataFrame({"calisanlar":["mehmet","fatma","veli","ali"],"grup":["Muhasebe","Mühendislik","Mühendislik","İK"]})
sozluk1
```

```python
sozluk2=pd.DataFrame({"calisanlar":["mehmet","ali","fatma","veli"],"ilk_giriş":["2019","2011","2010","2021"]})
sozluk2
```

```python
pd.merge(sozluk1,sozluk2)
```

```python
pd.merge(sozluk1,sozluk2,on="calisanlar") #iki veri setinde ortak polan grup hangisisyse ona göre birleştirme işlemi yapabilirz.
```

**Çoktan Teke**

```python
sozluk3=pd.merge(sozluk1,sozluk2)
sozluk3
```

```python
sozluk4=pd.DataFrame({"grup":["Muhasebe","Mühendislik","İK"],"mudur":["Caner","Mustafa","Berkcan"]})
sozluk4
```
