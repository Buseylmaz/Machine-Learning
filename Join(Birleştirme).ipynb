```python
import pandas as pd
import numpy as np
```

```python
a=np.random.randint(0,30,size=(4,3))
b1=pd.DataFrame(a,columns=("v1","v2","v3"))
b1
```

**Concat ile birleştirme**

```python
b2=b1+25
b2
```

```python
pd.concat([b1,b2] ,ignore_index=True) #index numarası 3 den sonra sırasıyla devam etsin diye
```

```python
b1.columns=["v1","v2","v4"]
b1
```

```python
pd.concat([b1,b2],ignore_index=True )
```

**Join**

```python
pd.concat([b1,b2],join="inner" )
```

```python
pd.concat([b1,b2])[b1.columns]
```
